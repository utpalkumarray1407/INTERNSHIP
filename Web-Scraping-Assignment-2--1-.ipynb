[QUESTION - 1] : 

Write a python program to scrape data for “Data Analyst” Job position in “Bangalore” location. You have to scrape the job-title, job-location, company_name, experience_required. You have to scrape first 10 jobs data. This task will be done in following steps: 

1. First get the webpage https://www.shine.com/ 

2. Enter “Data Analyst” in “Job title, Skills” field and enter “Bangalore” in “enter the  location” field. 

3. Then click the search button. 

4. Then scrape the data for the first 10 jobs results you get. 

5. Finally create a dataframe of the scraped data. 

Note: All of the above steps have to be done in code. No step is to be done manually

```python
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings("ignore")
from selenium.webdriver.common.by import By
import time
```

```python
driver = webdriver.Chrome()
```

```python
driver.get("https://www.shine.com/")
```

```python
designation=driver.find_element(By.CLASS_NAME,"form-control")
designation.send_keys("Data Analyst")
```

```python
location=driver.find_element(By.XPATH,"/html/body/div/div[4]/div/div[2]/div[2]/div/form/div/div[1]/ul/li[2]/div/input")
location.send_keys("Banglore")
```

```python
search=driver.find_element(By.CLASS_NAME,"searchForm_btnWrap_advance__VYBHN")
search.click()
```

```python
job_title = []
job_location = []
company_name = []
experience_required = []
```

```python
title_tags=driver.find_elements(By.XPATH,'//h2[@itemprop="name"]/a')
for i in title_tags:
    title=i.text
    job_title.append(title)
```

```python
location_tags=driver.find_elements(By.XPATH,'//div[@class=" jobCard_jobCard_lists_item__YxRkV jobCard_locationIcon__zrWt2"]')
for i in location_tags:
    location=i.text
    job_location.append(location)
```

```python
company_tags=driver.find_elements(By.XPATH,'//div[@class="jobCard_jobCard_cName__mYnow"]/span')
for i in company_tags:
    company=i.text
    company_name.append(company)
```

```python
experience_tags=driver.find_elements(By.XPATH,'//div[@class=" jobCard_jobCard_lists_item__YxRkV jobCard_jobIcon__3FB1t"]')
for i in experience_tags:
    experience=i.text
    experience_required.append(experience)
    
```

```python
print(len(job_title),len(job_location),len(company_name),len(experience_required))
```

```python
import pandas as pd
df=pd.DataFrame({'Title': job_title ,'Job Location': job_location ,'Company Name': company_name,'Experience Required': experience_required})
df
```

Write a python program to scrape data for Top 1000 Quotes of All Time.

```python
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings("ignore")
from selenium.webdriver.common.by import By
import time
```

```python
driver = webdriver.Chrome()
```

```python
driver.get("https://www.azquotes.com/")
```

```python
quote_line=[]
author_name=[]
quote_type=[]
```

```python
line_tags=driver.find_elements(By.XPATH,'//a[@class="title"]')
for i in line_tags:
    line=i.text
    quote_line.append(line)
```

```python
name_tags=driver.find_elements(By.XPATH,'//div[@class="author"]')
for i in name_tags:
    name=i.text
    author_name.append(name)
```

```python
type_tags=driver.find_elements(By.XPATH,'//div[@class="tags"]')
for i in type_tags:
    type=i.text
    quote_type.append(type)
```

```python
print(len(quote_line),len(author_name),len(quote_type))
```

```python
import pandas as pd
df = pd.DataFrame({ 'Quote': quote_line , 'Author': author_name , 'Type Of Quotes': quote_type})
```

```python
df
```

```python
quote_line=[]
author_name=[]
quote_type=[]
```

```python
start = 0
end = 10
for page in range(start,end):
    line=driver.find_elements(By.XPATH,'//a[@class="title"]')
    for i in line:
        quote_line.append(i.text)
    name=driver.find_elements(By.XPATH,'//div[@class="author"]')
    for i in name:
        author_name.append(i.text)
    type=driver.find_elements(By.XPATH,'//div[@class="tags"]')
    for i in type:
        quote_type.append(i.text)
```

```python
next_button=driver.find_element(By.XPATH,'/html/body/div[1]/div[2]/div/div/div/div[1]/div/div[4]/li[12]')
next_button.click()
time.sleep(10)
```

```python
print(len(quote_line),len(author_name),len(quote_type))
```

```python
import pandas as pd
df = pd.DataFrame({ 'Quote': quote_line , 'Author': author_name , 'Type Of Quotes': quote_type})
df
```
